```python

import matplotlib.pyplot as plt
import pandas as pd #trabajar con arrays arrchivos de datos
import pylab as pl
import numpy as np
```

```python
xls=pd.read_csv('Prueba2.0.csv')
x=np.linspace(0,119,120)   #espaciado cada 1, total de elementos=120
y=np.linspace(0,-21.91,220)    #espaciado cada 0.1 ,total de elementos=220
prueba1=xls[['ezz']]
pruebaezz1=np.array(prueba1).reshape(220,120)
#np.sum(pruebaezz1[:40,33])
```

```python
xls1=pd.read_csv('Prueba3.0.csv')

x1=np.linspace(0,159,160)   #espaciado cada 1, total de elementos=120
y1=np.linspace(0,-21.81,110)    #espaciado cada 0.1 ,total de elementos=220
prueba=xls1[['ezz']]
pruebaezz=np.array(prueba).reshape(110,160)
#np.sum(pruebaezz[:40,33])/1000000
```

```python
#Dezplazamiento vertical 40 kN 
uno=xls[['Uz']]
Uz=np.array(uno).reshape(220,120)
xmesh,ymesh=np.meshgrid(x,y)

contorno=plt.contourf(xmesh,ymesh,Uz)
c=plt.colorbar(contorno)
plt.title('Desplazamiento vertical Uz 45 kN')
plt.xlabel('Distancia [in]')
plt.ylabel('Profundidad [in]')
c.ax.set_ylabel('Uz [mils]')
plt.grid(c='k', ls='-', alpha=0.5)
plt.contour(xmesh,ymesh,Uz,colors='k')
plt.axhline(y=-3.937,c='r',linestyle='--')   #limite carpeta asfaltica
plt.axhline(y=-11.811,c='r',linestyle='--')  #limite base
plt.axhline(y=-19.685,c='r',linestyle='--')  #limite base
plt.show()


```

```python

#Dezplazamiento vertical 65 kN
uno1=xls1[['Uz']]
Uz1=np.array(uno1).reshape(110,160)
xmesh1,ymesh1=np.meshgrid(x1,y1)
contorno1=plt.contourf(xmesh1,ymesh1,Uz1)
c=plt.colorbar(contorno1)
plt.title('Desplazamiento vertical Uz 65 kN')
plt.xlabel('Distancia [in]')
plt.ylabel('Profundidad [in]')
c.ax.set_ylabel('Uz [mils]')
plt.grid(c='k', ls='-', alpha=0.5)
plt.contour(xmesh1,ymesh1,Uz1,colors='k')
plt.axhline(y=-3.937,c='r')   #limite carpeta asfaltica
plt.axhline(y=-11.811,c='r')  #limite base
plt.axhline(y=-19.685,c='r')  #limite base
plt.show()
```
